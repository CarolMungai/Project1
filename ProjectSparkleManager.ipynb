```python
### Sparkle Manager

chores = []

def menu_options():
    """Display the menu of available options."""
    print("Welcome to the Sparkle Manager!")
    print("1. Add chore")
    print("2. Mark chore as Completed")
    print("3. List chores")
    print("4. Remove chore")
    print("5. Quit")

def add_chore(chore_description):
    """Add a chore to the list of chores."""
    chores.append({"description": chore_description, "completed": False})
    print("chore added!")

def mark_chore_completed(chore_index):
    """Mark a chore as completed."""
    if 0 <= chore_index < len(chores):
        if chores[chore_index]["completed"]:
            print("chore is already marked as completed.")
        else:
            chores[chore_index]["completed"] = True
            print("chore marked as completed.")
    else:
        print("Invalid chore index.")

def list_chores():
    """List all chores along with their completion status."""
    print("chores:")
    for index, chore in enumerate(chores):
        status = "Completed" if chore["completed"] else "Not Completed"
        print(f"{index}. {chore['description']} - {status}")

def remove_chore(chore_index):
    """Remove a chore from the list."""
    if 0 <= chore_index < len(chores):
        removed_chore = chores.pop(chore_index)
        print(f"chore '{removed_chore['description']}' removed.")
    else:
        print("Invalid chore index.")

def main():
    """Main function to run the Sparkle Manager."""
    while True:
        menu_options()
        choice = input("Select an action: ")

        if choice == "1":
            chore_description = input("Enter chore description: ")
            add_chore(chore_description)
        elif choice == "2":
            chore_index = int(input("Enter chore index to mark as completed: "))
            mark_chore_completed(chore_index)
        elif choice == "3":
            list_chores()
        elif choice == "4":
            chore_index = int(input("Enter chore index to remove: "))
            remove_chore(chore_index)
        elif choice == "5":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
